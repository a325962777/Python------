```python
years=[2000,2001,2005,2010,2020]
years2=[100,1100,1200,1300,1400]
```

```python
energy=['a','b','c','d','e']
```

```python
plt.plot(years,energy)
plt.plot(years2,energy)
```

```python
fig, axs = plt.subplots(2, 2)  # יוצר מטריצה של גרפים בגודל 2x2
axs[0, 0].plot(years,energy, 'tab:blue')
axs[0, 0].set_title('a')
axs[0, 1].plot(years,energy ,'tab:red')
axs[0, 1].set_title('b')
axs[1, 0].scatter(years,energy, c='yellow')
axs[1, 0].set_title('c')
axs[1, 1].scatter(years,energy, c='green')
axs[1, 1].set_title('d')

# התאמת הגרפים למראה נקי יותר
for ax in axs.flat:
    ax.label_outer()
```
